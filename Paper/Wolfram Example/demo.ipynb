```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-10, 10, 500)
y = np.arcsinh(x)

plt.plot(x, y, label="arcsinh(x)")
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.grid(True)
plt.xlabel("x")
plt.ylabel("arcsinh(x)")
plt.title("ArcSinh[x]")
plt.legend()
plt.show()
```

```python
Plot[
  Log[Abs[x + Sqrt[x^2 - 1]]],
  {x, 1, 10},
  PlotStyle -> Blue,
  PlotRange -> {{0, 10}, {-1, 3}},
  AxesOrigin -> {0, 0},
  PlotLabel -> Row[{
    TraditionalForm[ln Abs[x + Sqrt[x^2 - 1]]],
    ",  x \[Element] (1, \[Infinity])"
  }],
  LabelStyle -> {FontSize -> 35},
  ImageSize -> 900
]
```

```python
Plot[
  -Log[Abs[x + Sqrt[x^2 - 1]]],
  {x, -10, -1},
  PlotStyle -> Red,
  PlotRange -> {{-10, 0}, {-1, 3}},
  AxesOrigin -> {0, 0},
  PlotLabel -> Row[{
    TraditionalForm[-ln Abs[x + Sqrt[x^2 - 1]]],
    ",  x \[Element] (-\[Infinity], -1)"
  }],
  LabelStyle -> {FontSize -> 35},
  ImageSize -> 900
]
```

```python
(* Normal definite integral *)
normalExact = Integrate[1/Sqrt[x^2 - 1], {x, 2, 3}]
normalDecimal = N[normalExact]

(* Improper definite integral *)
improperExact = Integrate[1/Sqrt[x^2 - 1], {x, 1, 2}]
improperDecimal = N[improperExact]
```
